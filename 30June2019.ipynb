```python
#Function to read Csv data into a DataFrame and  
#Returns the DataFrame Object

filepath="DataFiles/Income.csv"
import pandas as pd
def readCSVdata(filepath):
    return pd.read_csv(filepath)
readCSVdata(filepath)
```

```python
incomepdf = readCSVdata(filepath)
# Function to print all columns names in a singe line04000US06
#GEOID	State	2005	2006	2007	2008	2009	2010	2011	2012	2013-->output

def printDataframeColumns(df):
    columns=df.columns # Method
    #print(columns)
    for coloumn in columns:
        print(coloumn,end=" ")
    return
printDataframeColumns(incomepdf)
```

```python
#Function to access a row based on a unique column value
def AccessDataFrameRow(df,key):
    #print(df.values)
    for row in df.values:
        
        if key in row:
            for item in row:
                print(item,end=" ")
                #print(type(item))
    return

AccessDataFrameRow(incomepdf,57848)
```

```python
#Accessing a unique value based on row,column information
#Income of State for a particular Year
def valueFromRowColumn(df,rowkey,columnkey):
    for i in range(len(df.values)):
        if df.values[i][0]==rowkey or df.values[i][1]==rowkey:
            rowindex=i
    for i in range(len(df.columns)):
        if df.columns[i]== columnkey:
            columnindex = i
    return df.values[rowindex][columnindex]
            
    #return rowindex
valueFromRowColumn(incomepdf,"Alaska","2009")
```

```python
def getRowindex(df,rowkey):
    for i in range(len(df.values)):
        if df.values[i][0]==rowkey or df.values[i][1]==rowkey:
            rowindex=i
    return rowindex
def getColumnindex(df,columnkey):
    #rowindex=getRowindex(df,rowkey)
    for i in range(len(df.columns)):
        if df.columns[i]== columnkey:
            columnindex = i
    return columnindex
def valueFromRowColumn(df,rowkey,columnkey):
    rowindex=getRowindex(df,rowkey)
    columnindex=getColumnindex(df,columnkey)
    return df.values[rowindex][columnindex]

    
valueFromRowColumn(incomepdf,"Alaska","2009")
```

```python
#Function to update data based on rowkey and colunmkey
def updateDataFromRowColumn(df, rowkey ,columnkey, newdata):
    rowindex = getRowindex(df, rowkey)
    print(rowindex)
    columnindex = getColumnindex(df,  columnkey)
    print(columnindex)
    row=df.values[rowindex]
    print(row)
    row[columnindex]=newdata
    df.loc[rowindex]=row
    return

updateDataFromRowColumn(incomepdf, 'Arizona', '2007', "629")
incomepdf
```
